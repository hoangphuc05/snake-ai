```python
# https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/
# import Pandas and Keras NN library
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Read in the simple data set
data = pd.read_csv("covid_data.csv", delimiter=',')
```

```python
data
```

```python
data.columns
```

```python
# remove special character
data.columns = data.columns.str.replace(' ', '')
```

```python
# Get the X and Y data from the data set
X = pd.concat( [ data['VotedforTrump'], data['VotedforBiden'] ], axis = 1 )
Y = pd.get_dummies(data[['Restrictions']])  # Need to change to dummy variables
```

```python
X
```

```python
Y
```

```python
# Make the Neural Network Model
model = Sequential()
model.add(Dense(3, input_dim=2, activation='sigmoid'))
model.add(Dense(3, activation='sigmoid'))
model.add(Dense(2, activation='sigmoid'))
```

```python
# Compile the Keras Model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the keras model on the dataset
model.fit(X,Y, epochs=150, batch_size=10)
```

```python
# Evaluate the keras model
model.evaluate(X,Y)
```
