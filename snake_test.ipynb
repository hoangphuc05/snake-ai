```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Read in the simple data set
data = pd.read_csv("snake_game_data.csv", delimiter=',')
```

```python
data
```

```python
# remove special character
data.columns = data.columns.str.replace(' ', '')
data.columns
```

```python
X = pd.concat( [data['foodDiffX'], data['foodDiffY'], data['up_collision'], data['down_collision'], data['left_collision'], data['right_collision'], data['direction']], axis=1)
Y = pd.get_dummies(data['action'])
```

```python
X
```

```python
Y
```

```python
# Make the Neural Network Model
model = Sequential()
model.add(Dense(7, input_dim=7, activation='relu'))
model.add(Dense(14, activation='relu'))
model.add(Dense(14, activation='relu'))
model.add(Dense(4, activation='relu'))
```

```python
# Compile the Keras Model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the keras model on the dataset
model.fit(X,Y, epochs=150, batch_size=20)
```

```python
# Evaluate the keras model
model.evaluate(X,Y)
```

```python
model.summary()
```

```python
model.predict(pd.DataFrame([[-100.0,-20.0,13,7,12,3]]))
```
